```python
import pandas as pd
```

We are now going to read in a CSV into a DataFrame. A DataFrame is similar to a table or worksheet in Excel. 

Reading a CSV can be done by executing `pd.read_csv("path_to_your_file.csv")`

Run the cell below and notice the following: 
1. The data is formatted as a table (similar to excel)
2. We have an index column (the first column without a name and starts from 0) which is similar to the row number in excel
3. We have column headers (similar to excel)

```python
# read a CSV into a DataFrame
pd.read_csv("https://raw.githubusercontent.com/AustralianSDAF/python-data-visualisation/main/perth_house_prices.csv")
```

```python
# we often like to read store the DataFrame as a variable so that we can access it later
df = pd.read_csv("https://raw.githubusercontent.com/AustralianSDAF/python-data-visualisation/main/perth_house_prices.csv")
```

```python
# show  the first 5 rows
df.head(5)
```

```python
# show the last 5 rows
df.tail(5)
```

```python
# selecting only "BEDROOMS", "SUBURB" and "PRICES" columns 
df[["SUBURB", "BEDROOMS", "PRICE"]].head(5)
```

```python
# creating a new column: "free_space" = "LAND_AREA" - "FLOOR_AREA"
df["free_space"] = df["LAND_AREA"] - df["FLOOR_AREA"]
df[["free_space", "LAND_AREA", "FLOOR_AREA"]].head(5)
```

```python
# creating a new column called "AGE" = 2021 - "BUILD_YEAR"
df["age"] = 2021 - df["BUILD_YEAR"]
df[["age", "BUILD_YEAR"]].head(5)
```

# Filtering

The below statement is simply passing a Series of True/False objects to the DataFrame, returning all rows with True.

```python
# similarly, we can filter our DataFrame by doing: 

df[df["age"] < 5] # filters where age < 5
```

```python
# we can filter with multiple conditions by doing 
df[(df["age"] < 10) & (df["BEDROOMS"]==4)] # filters where age <> 10 and BEDROOMS == 4
```

# If/then logic
Let’s say we want to make a comment column with values of low and high, based on whether the PRICE is less or more than $1000000.

```python
import numpy as np 
```

```python
# perform the if/then statement by using np.where()
df["comment"] = np.where(df["PRICE"] < 1000000, "low", "high")
df.head(5)
```

```python
# we can drop a column by doing
df.drop("NEAREST_SCH_RANK", axis=1) # axis=1 tells pandas to drop "NEAREST_SCH_RANK" from the columns
```

```python
# we can rename a column by passing in a key-value pair that contains: {'from': 'to'}
df = df.rename(columns={"NEAREST_SCH_DIST": "Dist_to_school"})
df.head(5)
```

```python
# finally, we can sort a DataFrame by doing
df.sort_values(["PRICE"]).head(10) # sort by lowest to highest bill by default
```

```python
df.sort_values(["PRICE"], ascending=False).head(10) # sort by highest to lowest
```

```python
# obtain the count of the number of houses in bentley
bentley_houses = len(df[df["SUBURB"] == "Bentley"])

# obtain the count of the number of houses in Peppermint Grove
peppermint_houses = len(df[df["SUBURB"] == "Peppermint Grove"])

# obtain the count of the number of rows that don't have data on ranking of nearest school
no_school_data = len(df[df["NEAREST_SCH_RANK"].isna()])

print(f"The number of houses in bentley are: {bentley_houses}")
print(f"The number of houses in peppermint grove are: {peppermint_houses}")
print(f"The number of houses that don't have data on nearest school ranking: {no_school_data}")
```

```python
# get the unique list of suburb using .unique()
df["SUBURB"].unique()
```

```python
# get the unique list of SUBURBS that are within the boundary of Perth Modern School using .unique()
df[df["NEAREST_SCH"]=="PERTH MODERN SCHOOL"]["SUBURB"].unique()
```

```python
# sort the houses that are less than 800k by nearest school ranking.
df[df["PRICE"] < 800000].sort_values(["NEAREST_SCH_RANK"], ascending=True).head(50)
```
