## Task 1: 

Write a NumPy program to create a 3X4 matrix and iterate over it and print each element.

```python
# write your program here
a = np.arange(10,22).reshape((3, 4))
print("Original array:")
print(a, end="\n\n")

print("Each element of the array is:")
for x in np.nditer(a):
  print(x)
```

## Task 3: 

Write a NumPy program to calculate the difference between the maximum and the minimum values of a given array along the second axis

```python
x = np.arange(12).reshape((2, 6))
print("\nOriginal array:")
print(x)

min_value = np.min(x,axis=1)
print(min_value)

max_value = np.max(x,axis=1)
print(max_value)

difference = max_value - min_value
print(f"Difference between the maximum and the minimum values of the said array: {difference}")
```
