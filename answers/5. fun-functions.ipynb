## Task 1: 

Question: Does the age of a player affect a player's ratings? 

To answer this question, find the average age of the top and bottom 25 players in the world. 

How? 
1. Create a function that takes in a list of players as the input, and returns the average age of all players in the supplied list 
2. Inside of the function, you'll create a for loop to iterate through all players in the list 
3. Inside of the loop,  you'll sum all the player's age 
4. Outside of the loop, you'll get the total number of players by taking length of the list using the `len()` function: https://www.w3schools.com/python/ref_func_len.asp
5. Outside of the loop, calculate the average by taking `Sum of all players ages` / `Total number of players` 
6. Finally, invoke your function by passing in a list of all players

```python
# create your function here
def get_average_age(players):
    sum_of_ages = 0 
    for player in players:
        sum_of_ages = sum_of_ages + player["Age"]
    
    return sum_of_ages / len(players)
```

```python
# get the average age of the top 25 players and print the result 
average_age_top_25_players = get_average_age(top_25_players)
print(f"The average age of the top 25 players is: {average_age_top_25_players}")

# get the average age of the bottom 25 players and print the result 
average_age_bottom_25_players = get_average_age(bottom_25_players)
print(f"The average age of the bottom 25 players is: {average_age_bottom_25_players}")
```

## Task 2: 

Question: What are the best and worst clubs? 

To answer this question, find the average rating of clubs for the top 25 and bottom 25 players of the world. 

How? 
1. Like before, create a function that accepts a list of players 
2. Create a new dictionary to calculate the `total rating` for each club. You'll need to create new key-value pairs if the key does not exist. 
3. Create a new dictionary to calculate the `total number of players` for each club. You'll need to create new key-value pairs if the key does not exist. 
4. Create a new dictionary to store the average club rating by taking `total rating` / `total number of players` in each club. 

Hint: This question requires the use of `.get()` on a dictionary. 

```python
# create your function here
def get_club_average_rating(players):
    club_total_rating = {}
    club_total_players = {}

    for player in players:
        if club_total_rating.get(player["Club"]) is None:
            club_total_rating[player["Club"]] = player["Rating"]
            club_total_players[player["Club"]] = 1
        
        else: 
            club_total_rating[player["Club"]] = club_total_rating[player["Club"]] + player["Rating"]
            club_total_players[player["Club"]] = club_total_players[player["Club"]] + 1

    club_average_rating = {}

    for club in club_total_rating:
        if club_average_rating.get(club) is None: 
            club_average_rating[club] = club_total_rating[club] / club_total_players[club]

    return club_average_rating
```

```python
# get the average rating for the top clubs and print the result 
top_clubs_average_rating = get_club_average_rating(top_25_players)
print(f"The average rating for each of the top clubs are: {top_clubs_average_rating}")

# get the average rating for the bottom clubs and print the result
bottom_clubs_average_rating = get_club_average_rating(bottom_25_players)
print(f"The average rating for each of the bottom clubs are: {bottom_clubs_average_rating}")
```
